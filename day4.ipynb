```python
dcs = []
with open("input4.txt", "r") as f:
    accu = {}
    for line in f.readlines():
        if line == "\n":
            dcs.append(accu)
            accu = {}
            continue
        for a in line.split(" "):
            b, c = a.split(":")
            accu[b] = c.rstrip("\n")
    dcs.append(accu)
```

```python
needed = ["byr","iyr","eyr","hgt","hcl","ecl","pid"]

def check(dc):
    ks = list(dc.keys())
    for x in needed:
        if not x in ks:
            return False
    return True
```

```python
cnt = 0
for x in dcs:
    if check(x):
        cnt += 1
print(cnt)
```
