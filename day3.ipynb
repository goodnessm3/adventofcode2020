```python
myarr = []
with open("input3.txt", "r") as f:
    for line in f.readlines():
        myarr.append(line.rstrip("\n"))
```

```python
myarr
```

```python
x = 0
y = 0
cnt = 0
maxlen = len(myarr[0])
while True:
    print(x,y)
    if myarr[x][y] == "#":
        cnt += 1
    x += 1
    y += 3
    if y >= maxlen:
        y -= maxlen
        
```

```python
cnt
```

```python
def general(right, down):
    x = 0
    y = 0
    cnt = 0
    maxlen = len(myarr[0])
    while True:
        try:
            if myarr[x][y] == "#":
                cnt += 1
        except IndexError:
            return cnt
        x += down
        y += right
        if y >= maxlen:
            y -= maxlen
```

```python
general(3,1)
```

```python
slopes = [(1,1),(3,1),(5,1),(7,1),(1,2)]
results = []
for sl in slopes:
    results.append(general(*sl))
```
